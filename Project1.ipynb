```python
#Step 3
r = requests.get("http://cdaw.gsfc.nasa.gov/CME_list/radio/waves_type2.html",headers=headers)
bs = BeautifulSoup(r.content,"html.parser")
bs.prettify()
table = bs.find("pre")
table = table.get_text()
table = table.split("\n")[12:494]
wave_df= pd.DataFrame(table)

temp = lambda row : pd.Series(row.split())
wave_df = wave_df[0].apply(temp)
wave_df = wave_df.drop([15, 16, 17, 18, 19, 20, 21, 22, 23], axis =1)
wave_df.columns = ["start_date", "start_time", "end_date", "end_time", "start_frequency", "end_frequency", "flare_location", "flare_region","flare_classification", "cme_date", "cme_time", "cme_angle", "cme_width", "cme_speed", "plot_type"]
wave_df
```

After extracting the data from the web, I find the table and slice it to keep only the needed informtaion. Then I use the annonymous function to split the data into columns and drop the blank columns. In the end I name each column in the table.

```python
#Step 4
wave_df.replace(['----', '-----', '--/--', '--:--', '????', 'FILA', 'DSF','EP', 'DIM', 'EP?' ],'NaN', inplace = True )

wave_df['is_halo']= np.where(wave_df['cme_angle'] == 'Halo', 'True', 'False')
wave_df.replace('Halo', 'NaN', inplace = True)

wave_df['width_lower_bound'] = wave_df['cme_width'].map(lambda x: str(x)[0] == '>')
wave_df['cme_width'] = wave_df['cme_width'].str.replace('>','')
wave_df['end_time'].replace('24:00', '00:00',inplace = True) 
wave_df
```

I replace the unidentify data by NaN and then create a new column named is_halo indicates if the flare is halo or not using the cme_angle column. After that, I create a new column called width_lower_bound using annonymous function. finally, I reformat the time presentation in case it causes some errors  later on. 

```python
                                     
from datetime import datetime

for _,col in wave_df.iterrows():
    year = col['start_date'][:5]

    col['start_date'] = col['start_date']+' '+col['start_time']
    col['start_date'] = datetime.strptime(col['start_date'], '%Y/%m/%d %M:%S')

    if pd.notna(col['end_date']) and pd.notna(col['end_time']):
        col['end_date'] = str(year) + str(col['end_date'])+' '+str(col['end_time'])
        col['end_date'] = datetime.strptime(col['end_date'], '%Y/%m/%d %M:%S' )
    if col['cme_date'] != 'NaN' and col['cme_time'] != 'NaN':
        col['cme_date']= str(year)+str(col['cme_date'])+' '+str(col['cme_time'])
        col['cme_date']= datetime.strptime(col['cme_date'], '%Y/%m/%d %M:%S')


wave_df
```

```python
#Part2
#Question 1:
top_50 = wave_df.loc[wave_df['flare_classification'].astype(str).str.contains('X')].copy()
top_50['top_50_solar_flare'] = top_50['flare_classification']
top_50['top_50_solar_flare'] = top_50['top_50_solar_flare'].str.replace('X','')
top_50['top_50_solar_flare'] = top_50['top_50_solar_flare'].astype(float)

top_50 = top_50.sort_values('top_50_solar_flare', ascending = False)
top_50.drop('top_50_solar_flare', axis = 1, inplace = True)

top_50 = top_50.head(50)

top_50
```

By looking at the expected table in part 1, Step 4, Top 50 table and the wave flare table have 2 common fields, which is the flare classification and flare region. We can iterate through those two columns of both tables to find the best matching row. 


```python
top_50['rank'] = 0

for i, row_df in df.iterrows():
    for j, row_50 in top_50.iterrows():
        if row_df['region'] == int(row_50['flare_region']) and row_df['x_class'] == row_50['flare_classification']:
            top_50.at[j, 'rank']= row_df['rank']

```
