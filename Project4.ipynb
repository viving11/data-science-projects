```python
#dradient descent for multivariable linear regression
def grad_descent(X,y,T,alpha):
    m, n = X.shape
    theta = np.zeros(n)
    f = np.zeros(T)
    for i in range(T):
        f[i] = 0.5*np.linalg.norm(X.dot(theta)-y)**2 #loss function
        g = np.transpose(X).dot(X.dot(theta)-y) #gradient
        theta = theta - alpha*g #theta
    return(theta,f)
```
